```python
import pandas as pd 
import numpy as np
import random
import statistics
```

```python
data = []

for i in range(8):
  num1 = random.randint(0,5)
  array1 = [num1, num1, 0, 0, "straight"]
  num2 = random.randint(0,5)
  array2 = [0,0,num2,num2, "straight"]
  data.append(array1)
  data.append(array2)

for i in range(8):
  num1 = random.randint(0,5)
  array1 = [num1, 0, 0,  num1, "diagonal"]
  num2 = random.randint(0,5)
  array2 = [0, num2, num2, 0, "diagonal"]
  data.append(array1)
  data.append(array2)
```

```python
df = pd.DataFrame(data, columns=["Top Left", "Top Right", "Bottom Left", "Bottom Right", "Class"])
df
```

```python
def sfunc(index1):
  listl = []
  thing = df[df.columns[index1]]
  thingy = thing[0:16]
  for pixel in thingy:
    listl.append(pixel)
  return listl

def dfunc(index2):
  listy = []
  anotherthing = df[df.columns[index2]]
  thingyy = anotherthing[16:32]
  for pixel in thingyy:
    listy.append(pixel)
  return listy


straight_top_left = sfunc(0)
straight_top_right = sfunc(1)
straight_bottom_left = sfunc(2)
straight_bottom_right = sfunc(3)

diagonal_top_left = dfunc(0)
diagonal_top_right = dfunc(1)
diagonal_bottom_left = dfunc(2)
diagonal_bottom_right = dfunc(3)
```

```python
print("Straight")
print(statistics.mean(straight_top_left[0:6]))
print(statistics.mean(straight_top_right[0:6]))
print(statistics.mean(straight_bottom_left[0:6]))
print(statistics.mean(straight_bottom_right[0:6]))
print("Diagonal")
print(statistics.mean(diagonal_top_left[0:6]))
print(statistics.mean(diagonal_top_right[0:6]))
print(statistics.mean(diagonal_bottom_left[0:6]))
print(statistics.mean(diagonal_bottom_right[0:6]))
```
