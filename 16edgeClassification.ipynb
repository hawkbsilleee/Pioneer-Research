```python
import pandas as pd 
import random

csv_file = "/content/drive/MyDrive/Pioneer /IRIS.csv"

df = pd.read_csv(csv_file)
```

```python
df
```

```python
def convert_to_list(index):
  return df.loc[index, :].values.tolist()
```

```python
setosa = []
versicolor = []
virginica = []
for i in range(10):
  index = int(random.random()*48)
  setosa.append(convert_to_list(index))

for i in range(10):
  index = int(random.randint(50, 100))
  versicolor.append(convert_to_list(index))

for i in range(10): 
  index = int(random.randint(101, 149))
  virginica.append(convert_to_list(index))

print(setosa)
print(versicolor)
print(virginica)
```

```python
traindata = pd.DataFrame(setosa, columns = ['sepal_length', "sepal_width", "petal_length", "petal_width", "species"])
for i in range(10):
  traindata.loc[len(traindata.index)] = versicolor[i]
for i in range(10):
  traindata.loc[len(traindata.index)] = virginica[i]
traindata
```

```python
globalmin = []
globalmin.append(traindata["sepal_length"].min())
globalmin.append(traindata["sepal_width"].min())
globalmin.append(traindata["petal_length"].min())
globalmin.append(traindata["petal_width"].min())

globalmax = []
globalmax.append(traindata["sepal_length"].max())
globalmax.append(traindata["sepal_width"].max())
globalmax.append(traindata["petal_length"].max())
globalmax.append(traindata["petal_width"].max())

minimum = min(globalmin)
maximum = max(globalmax) 
```

```python
def normalize(peramname): 
  min = traindata[peramname].min()
  max = traindata[peramname].max()
  x = traindata[peramname]
  num = 5 * ((x - minimum)/(maximum-minimum))
  return round(num, 2)  

traindata['sepal_length'] = normalize('sepal_length') 
traindata['sepal_width'] = normalize('sepal_width') 
traindata['petal_length'] = normalize('petal_length')
traindata['petal_width'] = normalize('petal_width')

traindata
```

```python
setosa_sepal_length = []
setosa_sepal_width = []
setosa_petal_length = []
setosa_petal_width = []
for i in range(10):
  setosa_list = []
  for thing in traindata.loc[i]:
    if thing != "Iris-setosa":
      setosa_list.append(thing)
  setosa_sepal_length.append(setosa_list[0])
  setosa_sepal_width.append(setosa_list[1])
  setosa_petal_length.append(setosa_list[2])
  setosa_petal_width.append(setosa_list[3])
```

```python
versicolor_sepal_length = []
versicolor_sepal_width = []
versicolor_petal_length = []
versicolor_petal_width = []
for i in range(10, 20):
  versicolor_list = []
  for thing in traindata.loc[i]:
    if thing != "Iris-versicolor":
      versicolor_list.append(thing)
  versicolor_sepal_length.append(versicolor_list[0])
  versicolor_sepal_width.append(versicolor_list[1])
  versicolor_petal_length.append(versicolor_list[2])
  versicolor_petal_width.append(versicolor_list[3])
```

```python
virginica_sepal_length = []
virginica_sepal_width = []
virginica_petal_length = []
virginica_petal_width = []
for i in range(20, 30):
  virginica_list = []
  for thing in traindata.loc[i]:
    if thing != "Iris-virginica":
      virginica_list.append(thing)
  virginica_sepal_length.append(virginica_list[0])
  virginica_sepal_width.append(virginica_list[1])
  virginica_petal_length.append(virginica_list[2])
  virginica_petal_width.append(virginica_list[3])
```

```python
import statistics

print("Setosa")
print(statistics.mean(setosa_sepal_length))
print(statistics.mean(setosa_sepal_width))
print(statistics.mean(setosa_petal_length))
print(statistics.mean(setosa_petal_width))
print(" ")
print("Versicolor")
print(statistics.mean(versicolor_sepal_length))
print(statistics.mean(versicolor_sepal_width))
print(statistics.mean(versicolor_petal_length))
print(statistics.mean(versicolor_petal_width))
print(" ")
print("Iris-virginic")
print(statistics.mean(virginica_sepal_length))
print(statistics.mean(virginica_sepal_width))
print(statistics.mean(virginica_petal_length))
print(statistics.mean(virginica_petal_width))
```
